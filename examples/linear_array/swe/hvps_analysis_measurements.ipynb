# Single TX Analysis 2 OEMs Rail 1

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

id = 100
datafile = "hvps_measurements/hvps_push_id_" + str(id) + ".npy"
data = np.load(datafile)
data = np.squeeze(data)
print(data.shape)
n_oems, n_polarities, n_level, n_units, n_samples = data.shape
# n_oems: {0, 1}
# n_polarities: 0: HVM, 1: HVP
# n_level: 0: rail 0, 1: rail 1
#n_units: 0: voltage, 1: current

fig, axs = plt.subplots(nrows=2, ncols=4, figsize=(24, 12))
V_lim = [0, 100]
I_lim = [0, 35]

t = np.arange(data.shape[-1])
t = t*2

## OEM0, HVP1
axs[0][0].plot(t, data[0, 1, 1, 0, :], 'blue')
axs[0][0].set_title('OEM0, HVP_1 Voltage')
axs[0][0].set_xlabel('Time [us]')
axs[0][0].set_ylabel('Voltage [V]')
axs[0][0].set_ylim(V_lim)

axs[1][0].plot(t, data[0, 1, 1, 1, :], 'orange')
axs[1][0].set_title('OEM0, HVP_1 Current')
axs[1][0].set_xlabel('Time [us]')
axs[1][0].set_ylabel('Current [A]')
axs[1][0].set_ylim(I_lim)

## OEM0, HVM1
axs[0][1].plot(t, data[0, 0, 1, 0, :], 'blue')
axs[0][1].set_title('OEM0, HVM_1 Voltage')
axs[0][1].set_xlabel('Time [us]')
axs[0][1].set_ylabel('Voltage [V]')
axs[0][1].set_ylim(V_lim)

axs[1][1].plot(t, data[0, 0, 1, 1, :], 'orange')
axs[1][1].set_title('OEM0, HVM_0 Current')
axs[1][1].set_xlabel('Time [us]')
axs[1][1].set_ylabel('Current [A]')
axs[1][1].set_ylim(I_lim)

## OEM1, HVP1
axs[0][2].plot(t, data[1, 1, 1, 0, :], 'blue')
axs[0][2].set_title('OEM1, HVP_1 Voltage')
axs[0][2].set_xlabel('Time [us]')
axs[0][2].set_ylabel('Voltage [V]')
axs[0][2].set_ylim(V_lim)

axs[1][2].plot(t, data[1, 1, 1, 1, :], 'orange')
axs[1][2].set_title('OEM1, HVP_1 Current')
axs[1][2].set_xlabel('Time [us]')
axs[1][2].set_ylabel('Current [A]')
axs[1][2].set_ylim(I_lim)

## OEM1, HVM1
axs[0][3].plot(t, data[1, 0, 1, 0, :], 'blue')
axs[0][3].set_title('OEM1, HVM_1 Voltage')
axs[0][3].set_xlabel('Time [us]')
axs[0][3].set_ylabel('Voltage [V]')
axs[0][3].set_ylim(V_lim)

axs[1][3].plot(t, data[0, 0, 1, 1, :], 'orange')
axs[1][3].set_title('OEM1, HVM_1 Current')
axs[1][3].set_xlabel('Time [us]')
axs[1][3].set_ylabel('Current [A]')
axs[1][3].set_ylim(I_lim)

for i in range(2):
    for j in range(4):
        axs[i][j].grid()



```
